A DataFrame object has two axes: “axis 0” and “axis 1”:

axis 0: Wherever you see this -> it represents rows
axis 1: Wherever you see this -> it represents columns

Understanding the “inplace” parameter can help us a lot of time and memory!

When inplace = False -> which is the default, then the operation is performed and it returns a copy of the object. You then need to save it to something.

```python
temp=df.set_index('CustomerId')# here by Default inplace = False
temp
```

```python
temp.loc[15569892]
```


While , When inplace = True -> the data is modified in place, which means it will return nothing and the dataframe is now updated.

# df.loc vs df.iloc

- df.loc
loc is label-based, which means that we have to specify the name of the rows and columns that we need to filter out.

```python
df.loc[1:5]
```

```python
df[df['Age'] >= 45]
```

```python
df.loc[df.Age>=45]
```

```python
df.loc[0]                            # forst row details 
```

```python
df.loc[:,"Geography"]                  
```

```python
df.iloc[:,3:9]
```

df.iloc (i before the loc stands for indexing)
iloc is integer index-based , we have to specify rows and columns by their integer index.

```python
df.iloc[15:20,3:7]
```

```python
df.iloc[16][3]
```

```python
df.iloc[0:3]
```

```python
df.iloc[15:20,'Gender':]
```

# Data Manipulation Operations

```python
df = pd.read_csv(r"C:\Anil\Projects\BIA\Day5 - Pandas\Churn_Modelling.csv")
```

```python
df.sort_values('Balance')
```

```python
df.sort_values(['Balance','EstimatedSalary'],ascending=[True,False])
```

```python
df[(df['CreditScore'] > 700) & (df['CreditScore'] < 900)].sort_values(['CreditScore'],ascending=True)
```

```python
df[['Gender','NumOfProducts']].groupby('Gender').mean()
```

```python
df[['Gender','Geography','EstimatedSalary','Balance']].groupby(['Gender','Geography']).agg(['mean','count'])
```

```python
df.pivot_table(index='CreditScore',values=['Age','Balance'])
```
